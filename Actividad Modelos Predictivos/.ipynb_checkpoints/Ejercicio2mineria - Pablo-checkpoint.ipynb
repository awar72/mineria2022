Probamos el algoritmo de clasificación de Naive Bayes

```python
from sklearn.model_selection import train_test_split

X= data.drop(['comprar'], axis=1)
Y= data['comprar']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 1)
```

```python
from sklearn.naive_bayes import GaussianNB

modeloNB = GaussianNB()


modeloNB.fit(X_train, y_train)


y_pred = modeloNB.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Probamos con la funcion para seleccionar las mejores columnas para el modelo

```python
from sklearn.feature_selection import SelectKBest

best=SelectKBest(k=5)
X_new = best.fit_transform(X, Y)
X_new.shape
selected = best.get_support(indices=True)
print(X.columns[selected])
```

```python
used_features =X.columns[selected]
```

```python
X= data[used_features]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 1)

modelo_x = GaussianNB()


modelo_x.fit(X_train, y_train)


y_pred = modelo_x.predict(X_test)

matriz_x = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz_x)
```

```python
print(classification_report(y_test, y_pred))
```

Podemos ver que el modelo empeora, si ocupamos la función para seleccionar las mejores features para el modelo, de acuerdo a las metricas revisadas es mejor ocupar todas las variables del dataset, ya que nos da una mejor presición.

Ahora probaremos con el algoritmo de clasificación de los arboles de desición para clasificaciones

```python
from sklearn.tree import DecisionTreeClassifier

X2= data.drop(['comprar'], axis=1)

Y2 = data['comprar']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, Y2, test_size=0.2, random_state = 1)

modeloAD = DecisionTreeClassifier(max_depth = 10,max_leaf_nodes= 8, min_samples_split= 2, random_state=29)
#10,8,2

modeloAD.fit(X_train, y_train)

y_pred = modeloAD.predict(X_test)
```

```python

matriz_x = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz_x)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

params = {"max_depth": [2,3,4,5,6,7,8,9,10], 
          "min_samples_split": [2,3,4],
          "max_leaf_nodes": [4,5,6,8]}
grid = GridSearchCV(estimator=modeloAD, param_grid=params,cv=10)
_=grid.fit(X_train, y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```
